```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# sensor data
sensor_distances = np.array([200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100])
sensor_rawData = np.array([580, 445, 340, 280, 245, 230, 214, 211, 210, 206])
```

```python
# Plot the raw data
plt.plot(sensor_rawData, sensor_distances, 'bo', label='Raw Data')
```

```python
# Find the fit
a, b, c = np.polyfit(1/sensor_rawData, sensor_distances, 2)
```

```python
# Plot the fit
plt.plot(sensor_rawData, sensor_distances, 'bo', label='Raw Data')
plt.plot(np.linspace(180, 600, 1000), a*(1/np.linspace(180, 600, 1000))**2 + b*(1/np.linspace(180, 600, 1000))+c, 'r-', label='Fit')
plt.legend()
plt.show()

# Print the fit
print("== Fit ==")
print("a = " + str(round(a)))
print("b = " + str(round(b)))
print("c = " + str(round(c)))

print()


print("== Best Fit ==")
print("a = 128062")
print("b = -33892")
print("c =  2426")
```
